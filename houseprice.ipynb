```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

print("All libraries are ready to use!")
```

```python
x = np.array([[1000], [1500], [2000], [2500], [3000], [3500],[500]])
y = np.array([200000, 250000, 310000, 360000, 400000, 460000,1000000])
model = LinearRegression()
model.fit(x, y)
print("Model is ready to use!")
```

```python
prediction = model.predict([[2200]])
print(f"Predicted price for 2200 sq ft:${prediction[0]:,.2f}")
```

```python
plt.scatter(x,y,color='blue',label='Actual Data')
plt.plot(x,model.predict(x),color='red',label='Best fit Line')
plt.title('House Price Prediction')
plt.xlabel('House Size (sq ft)')
plt.ylabel('House Price')
plt.legend()
plt.show()
```
