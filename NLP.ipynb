```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN, Dense


# NLP part
sentences = ["I love programming", "Python is great", "I hate errors"]

# NLP part Tokenization and padding
tokenizer = Tokenizer(num_words=100)
tokenizer.fit_on_texts(sentences)
sequences = tokenizer.texts_to_sequences(sentences)
padded = pad_sequences(sequences, maxlen=5)

# Make RNN model part
model = Sequential([
    Embedding(100, 8, input_length=5),
    SimpleRNN(16),
    Dense(1, activation='sigmoid')
])
print("NLP pre-processing end and RNN model make")
```

```python
import torch
import torch.nn as nn

# Mock data (assuming 'padded' exists)
padded = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
input_data = torch.LongTensor(padded)
labels = torch.FloatTensor([1, 1, 0]).view(-1, 1)

class SimpleRNNModel(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim):
        super(SimpleRNNModel, self).__init__()
        # Use the arguments passed to __init__
        self.embedding = nn.Embedding(vocab_size, embed_dim)
        self.rnn = nn.RNN(embed_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.embedding(x)
        # Corrected method call from self.run to self.rnn
        out, h_n = self.rnn(x)
        # Get the hidden state of the last time step
        last_output = out[:, -1, :]
        x = self.fc(last_output)
        output = self.sigmoid(x)
        return output, h_n

# Parameters
vocab_size = 100
embed_dim = 8
hidden_dim = 16

# Initialize
model_pt = SimpleRNNModel(vocab_size, embed_dim, hidden_dim)
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model_pt.parameters(), lr=0.01)

print("PyTorch RNN model initialized successfully")
```
